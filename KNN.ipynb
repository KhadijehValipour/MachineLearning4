
  KNN : K  Nearest Neightbors

```python
width = np.concatenate((apple_width , banana_width , watermelon_width))
length = np.concatenate((apple_length ,banana_length , watermelon_length))

X_train=np.array([width , length]).T
X_train.shape
```

```python
X_train[:7]
```

```python
Y_apple = np.zeros(n ,dtype='int')

Y_apple
```

```python
Y_banana = np.ones(n , dtype='int')

Y_banana
```

```python
Y_watermelon = np.ones(n , dtype='int') *2

Y_watermelon
```

```python
Y_train = np.concatenate((Y_apple , Y_banana , Y_watermelon))

Y_train
```

```python
def euclidean_distance (x1 ,x2) :
    return np.sqrt(np.sum((x1-x2) ** 2))
```

```python
new_fruit = np.array([8,7])

distances =[]
for x in X_train :
    d=euclidean_distance(new_fruit , x)
    distances.append(d)

distances
```

```python
k=7
```

```python
np.argsort(distances)[0:k]
```

```python
nearest_Neightbors = np.argsort(distances)[0:k]
```

```python
Y_train[nearest_Neightbors]
```

```python
result = np.bincount(Y_train[nearest_Neightbors])

result
```

```python
np.argmax(result)
```

```python
if np.argmax(result) == 0 :
    print("üçé")
elif np.argmax(result) == 1 :
    print("üçå")
else :
    print("üçâ")
```
